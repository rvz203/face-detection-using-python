```python
import shutil
import os
import random
import xml.etree.ElementTree as ET

def create_train_test_dirs(base_dir, train_ratio=0.8):
    if not os.path.exists(base_dir):
        os.makedirs(base_dir)

    train_dir = os.path.join(base_dir, 'train')
    test_dir = os.path.join(base_dir, 'test')
    os.makedirs(train_dir, exist_ok=True)
    os.makedirs(test_dir, exist_ok=True)

    for sub_dir in ['masked', 'unmasked']:
        os.makedirs(os.path.join(train_dir, sub_dir), exist_ok=True)
        os.makedirs(os.path.join(test_dir, sub_dir), exist_ok=True)

    return train_dir, test_dir

def organize_data(images, annotations, train_dir, test_dir, train_ratio=0.8):
    masked_images = []
    unmasked_images = []

    for image, annotation in zip(images, annotations):
        try:
            tree = ET.parse(annotation)
            root = tree.getroot()
            mask_tag = root.find('mask')
            if mask_tag is not None and mask_tag.text == '1':
                masked_images.append(image)
            else:
                unmasked_images.append(image)
        except ET.ParseError:
            print(f"Error parsing {annotation}")
            continue

    random.shuffle(masked_images)
    random.shuffle(unmasked_images)

    train_masked = masked_images[:int(train_ratio * len(masked_images))]
    test_masked = masked_images[int(train_ratio * len(masked_images)):]

    train_unmasked = unmasked_images[:int(train_ratio * len(unmasked_images))]
    test_unmasked = unmasked_images[int(train_ratio * len(unmasked_images)):]

    for img in train_masked:
        shutil.copy(img, os.path.join(train_dir, 'masked'))
    for img in test_masked:
        shutil.copy(img, os.path.join(test_dir, 'masked'))

    for img in train_unmasked:
        shutil.copy(img, os.path.join(train_dir, 'unmasked'))
    for img in test_unmasked:
        shutil.copy(img, os.path.join(test_dir, 'unmasked'))

base_dir = 'dataset'
train_dir, test_dir = create_train_test_dirs(base_dir)
organize_data(train_images, train_annotations, train_dir, test_dir)
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.layers import Dense, Dropout, Flatten
from tensorflow.keras.models import Model

def create_mask_detector_model(input_shape):
    base_model = MobileNetV2(weights="imagenet", include_top=False, input_shape=input_shape)
    head_model = base_model.output
    head_model = Flatten()(head_model)
    head_model = Dense(128, activation="relu")(head_model)
    head_model = Dropout(0.5)(head_model)
    head_model = Dense(1, activation="sigmoid")(head_model)  # Change here
    model = Model(inputs=base_model.input, outputs=head_model)

    for layer in base_model.layers:
        layer.trainable = False

    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model

# Define image shape and create the model
input_shape = (224, 224, 3)
mask_detector_model = create_mask_detector_model(input_shape)

train_datagen = ImageDataGenerator(rescale=1.0/255.0, rotation_range=20, zoom_range=0.15,
                                   width_shift_range=0.2, height_shift_range=0.2, shear_range=0.15,
                                   horizontal_flip=True, fill_mode="nearest")

test_datagen = ImageDataGenerator(rescale=1.0/255.0)

# Define training and validation data generators
train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(224, 224),
    batch_size=32,
    class_mode='binary'  # Change here
)

test_generator = test_datagen.flow_from_directory(
    test_dir,
    target_size=(224, 224),
    batch_size=32,
    class_mode='binary'  # Change here
)

# Train the model
mask_detector_model.fit(
    train_generator,
    steps_per_epoch=len(train_generator),
    validation_data=test_generator,
    validation_steps=len(test_generator),
    epochs=20
)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
import numpy as np

def evaluate_model(model, test_generator):
    test_steps = len(test_generator)
    predictions = model.predict(test_generator, steps=test_steps)
    predicted_classes = (predictions > 0.5).astype(int).flatten()  # Change here
    true_classes = test_generator.classes
    class_labels = list(test_generator.class_indices.keys())

    report = classification_report(true_classes, predicted_classes, target_names=class_labels, labels=[0, 1])
    cm = confusion_matrix(true_classes, predicted_classes, labels=[0, 1])
    return report, cm

report, cm = evaluate_model(mask_detector_model, test_generator)
print(report)
print(cm)
```

```python
import os
import cv2
import xml.etree.ElementTree as ET
import numpy as np
import random
import shutil
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

# Load Images and Annotations
def load_data(images_folder, annotations_folder):
    images = []
    annotations = []

    for filename in os.listdir(images_folder):
        if filename.endswith('.jpg') or filename.endswith('.png'):
            image_path = os.path.join(images_folder, filename)
            annotation_path = os.path.join(annotations_folder, filename.replace('.jpg', '.xml').replace('.png', '.xml'))

            if os.path.exists(annotation_path):
                images.append(image_path)
                annotations.append(annotation_path)

    return images, annotations

images_folder = './images'
annotations_folder = './annotations'
images, annotations = load_data(images_folder, annotations_folder)

# Dataset Splitting
train_images, test_images, train_annotations, test_annotations = train_test_split(images, annotations, test_size=0.2, random_state=42)

# Create Directory Structure and Organize Data
def create_train_test_dirs(base_dir, train_ratio=0.8):
    if not os.path.exists(base_dir):
        os.makedirs(base_dir)

    train_dir = os.path.join(base_dir, 'train')
    test_dir = os.path.join(base_dir, 'test')
    os.makedirs(train_dir, exist_ok=True)
    os.makedirs(test_dir, exist_ok=True)

    for sub_dir in ['masked', 'unmasked']:
        os.makedirs(os.path.join(train_dir, sub_dir), exist_ok=True)
        os.makedirs(os.path.join(test_dir, sub_dir), exist_ok=True)

    return train_dir, test_dir

def organize_data(images, annotations, train_dir, test_dir, train_ratio=0.8):
    masked_images = []
    unmasked_images = []

    for image, annotation in zip(images, annotations):
        try:
            tree = ET.parse(annotation)
            root = tree.getroot()
            mask_tag = root.find('mask')
            if mask_tag is not None and mask_tag.text == '1':
                masked_images.append(image)
            else:
                unmasked_images.append(image)
        except ET.ParseError:
            print(f"Error parsing {annotation}")
            continue

    random.shuffle(masked_images)
    random.shuffle(unmasked_images)

    train_masked = masked_images[:int(train_ratio * len(masked_images))]
    test_masked = masked_images[int(train_ratio * len(masked_images)):]

    train_unmasked = unmasked_images[:int(train_ratio * len(unmasked_images))]
    test_unmasked = unmasked_images[int(train_ratio * len(unmasked_images)):]

    for img in train_masked:
        shutil.copy(img, os.path.join(train_dir, 'masked'))
    for img in test_masked:
        shutil.copy(img, os.path.join(test_dir, 'masked'))

    for img in train_unmasked:
        shutil.copy(img, os.path.join(train_dir, 'unmasked'))
    for img in test_unmasked:
        shutil.copy(img, os.path.join(test_dir, 'unmasked'))

base_dir = 'dataset'
train_dir, test_dir = create_train_test_dirs(base_dir)
organize_data(train_images, train_annotations, train_dir, test_dir)

# Face Detection
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

def detect_faces(image_path):
    img = cv2.imread(image_path)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))
    return faces, img

# Show Detected Faces
def show_detected_faces(images, num_faces_to_show=10):
    count = 0
    fig, axes = plt.subplots(2, 5, figsize=(15, 6))  # Create a 2x5 grid of subplots
    fig.subplots_adjust(hspace=0.3, wspace=0.3)

    for image_path in images:
        faces, img = detect_faces(image_path)
        for (x, y, w, h) in faces:
            if count >= num_faces_to_show:
                break

            ax = axes[count // 5, count % 5]
            face = img[y:y+h, x:x+w]
            ax.imshow(cv2.cvtColor(face, cv2.COLOR_BGR2RGB))
            ax.axis('off')
            count += 1

        if count >= num_faces_to_show:
            break

    plt.show()

# Display 10 detected faces from the training set
show_detected_faces(train_images, num_faces_to_show=10)
```

```python
import matplotlib.pyplot as plt

# Assuming `history` is the history object returned by `model.fit()`
history = mask_detector_model.fit(
    train_generator,
    steps_per_epoch=len(train_generator),
    validation_data=test_generator,
    validation_steps=len(test_generator),
    epochs=20
)

def plot_training_history(history):
    # Plot training & validation accuracy values
    plt.figure(figsize=(12, 4))
    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'])
    plt.plot(history.history['val_accuracy'])
    plt.title('Model Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend(['Train', 'Validation'], loc='upper left')

    # Plot training & validation loss values
    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'])
    plt.plot(history.history['val_loss'])
    plt.title('Model Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend(['Train', 'Validation'], loc='upper left')

    plt.show()

# Call the function to plot the training history
plot_training_history(history)
```

```python
def calculate_iou(pred_mask, true_mask):
    intersection = np.logical_and(true_mask, pred_mask).sum()
    union = np.logical_or(true_mask, pred_mask).sum()
    if union == 0:
        return 0  # Avoid division by zero
    return intersection / union

def calculate_miou(model, generator):
    ious = []
    for i in range(len(generator)):
        x, y_true = generator[i]
        y_pred = (model.predict(x) > 0.5).astype(int)  # Binarize predictions

        for j in range(len(y_true)):
            iou = calculate_iou(y_pred[j].squeeze(), y_true[j].squeeze())
            ious.append(iou)

    return np.mean(ious)

# Calculate mIoU for the test data
miou = calculate_miou(mask_detector_model, test_generator)
print(f'mIoU: {miou:.4f}')
```
